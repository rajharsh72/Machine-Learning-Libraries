# BINNING in PANDAS


### When dealing with continuous numeric data, it is often helpful to bin the data into multiple buckets for further analysis. There are several different terms for binning including bucketing, discrete binning and discretization. 
### Pandas supports these approaches using the cut and function.

```python
df_ages = pd.DataFrame({'age': np.random.randint(21, 51, 8)})
df_ages
```

```python
# This code creates a new column called age_bins that sets the x argument to the 
#  age column in df_ages and sets the bins argument to a list of bin edge values. 
#  The left bin edge will be exclusive and the right bin edge will be inclusive.
#  The bins will be for ages: (20, 29] (someone in their 20s), (30, 39], and (40, 49].

df_ages['age_bins'] = pd.cut(x= df_ages['age'], bins = [20, 29, 39, 49])
df_ages
```

```python
df_ages['age_bins'].unique()
```

```python
# This code creates a new column called age_by_decade with the same first 2 
# arguments as above, and a third argument of labels set to a list of values 
# that correspond to how the age values will be put in bins by decades.

df_ages['age_by_decade'] = pd.cut(x=df_ages['age'], bins=[20, 29, 39, 49], labels=['20s', '30s', '40s'])
df_ages
```

# 3. Apply Function
#####  Apply function can be used to perform pre-proccesing/data manipulation on your data both row wise and column wise.
##### It is faster ethod than simply using for loop over the database
##### Almost everytime if we need to iterate over the dataframe, then we use apply function

```python
# Accessing data row-wise
data.apply(lambda x: x)
```

```python
# Accessing first row
data.apply(lambda x: x[0])
```

```python
# Accessing first column
data.apply(lambda x: x[0], axis=1)
```

```python
# Access by column name
data.apply(lambda x: x['Item_Type'], axis=1)
```

```python
# We can also use apply to implement a function or a condition

#I want to clip all the prices whose value is greater than 150

def clip_by_price(price):
    if price>150:
        price = 150
        
    return price

data["Item_MRP"].apply(lambda x: clip_by_price(x))[:10]
```

```python
# Suppose i want to label encode any data from categorical data to discrete data
```

```python
data["Outlet_Location_Type"][:10]
```

```python
#Now suppose i want to encode 'Tier1' as '0', 'Tier2' as '1', 'Tier3' as '3'

def encode(city):
    if city=='Tier 1':
        label=0
    elif city == 'Tier 2':
        label=1
    else:
        label=2
    return label


data['Outlet_Location_Type'] = data['Outlet_Location_Type'].apply(encode)
data['Outlet_Location_Type'][:10]
```

# DATA AGGREGATION

### 1. groupby

```python
# NOW  WANT TO FIND WHAT IS THE MEAN PRICE FOR EACH ITEM TYPE
# WE CAN USE GROUPBY
# FIRST WE WILL GROUP THE DATA BY Item_Type COLUMN
```

```python
price_by_item = data.groupby('Item_Type')
price_by_item.first()
```

```python
# Now we find the mean of Item_MRP

price_by_item.Item_MRP.mean()
```

```python
#Applying groupby on multiple columns

multiple_group = data.groupby(["Item_Type", "Outlet_Type"])
multiple_group.first()
```

### How are Outlet_Sizes distributed based on the city type:: using crosstab fn

#### this fn is used to get the initial feel of the data. Here we can validate some basic hypothesis

#### for eg., in this case "Outlet_Location_Type" is expected to affect the "Outlet_Size" significantly. This can be tested using cross-tabulation as shown below

```python
#generate cross-tab of outlet size based on the outlet city

pd.crosstab(data["Outlet_Size"], data["Outlet_Location_Type"], margins = True)
```

## How are the sales changing per year?? pivottable

```python
# Pandas can be used to create MS Excel pivot table
```

```python
# The fun thing about pandas pivot_table is you can get another POV on your data with only one line of code.
```

```python
# Most of the pivot_table params use default values so the only mandatory params you must add are data and index

# data: dataframe
# index: is the column, grouper, array you'd like to group your data by.
# values (optional): is the column you'd like to aggregate. If you do not specify this then the function will aggregate all numeric columns
```

```python
pd.pivot_table(data, index=["Outlet_Establishment_Year"], values="Item_Outlet_Sales")
```

```python
# By default it calculates the mean of the values that are passed
```

```python
# Creating pivot table using multiple columns

pd.pivot_table(data, index=["Outlet_Establishment_Year", "Outlet_Location_Type", "Outlet_Size"], values="Item_Outlet_Sales")
```

```python
# Creating pivot table using multiple columns and applying different aggregate functions

pd.pivot_table(data, index=["Outlet_Establishment_Year", "Outlet_Location_Type", "Outlet_Size"], values="Item_Outlet_Sales", aggfunc = [np.mean, np.median, min, max, np.std])
```
