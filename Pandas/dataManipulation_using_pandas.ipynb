```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('bigmart_data.csv')
data.head()
```

```python
#find the count of null values in each column
data.isnull().sum()
```

```python
# dropping the null values
data = data.dropna(how='any')
data.isnull().sum()
```

```python
data.head()
```

#  1. Sorting Dataframes

### <ol>1. sort_values(): to sort pandas dataframe by one or more columns
###  <ol>2. sort_index(): to sort pandas dataframe by row index

```python
# sorting by Item_Weight

sorted_data = data.sort_values(by='Item_Weight')
sorted_data.head(10)
```

```python
''' sort_value() has multiple parameters like:
        1. ascending: takes boolean value, True for ascending order, False for descending order
        2. inplace: whether to do inplace sorting or not. Reflets the changes in the original dataframe
'''
data.sort_values(by='Item_Weight', ascending=False, inplace=True)
data.head(10)
```

```python
data = pd.read_csv('bigmart_data.csv')
data = data.dropna(how='any')
```

```python
#Sorting data based on multiple columns
#Sorts the data by firsdt sorting the first column name passed and then sorting the second column
data.sort_values(by=['Item_Weight', 'Item_MRP'], ascending=False).head(10)
```

## Sorting based on index, sort_index()

```python
data.sort_index(inplace=True)
data.head(10)
```
