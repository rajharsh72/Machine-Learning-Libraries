```python
# ‘fillna()’ function helps to fill missing values in a Pandas Dataframe. 
# It is used for updating missing values with the overall mean/mode/median 
# of the column.
```

```python
import pandas as pd
from scipy.stats import mode
```

```python
data = pd.read_csv('bigmart_data.csv')
data.head()
```

```python
# Find the sum of null values for each column
data.isnull().sum()
```

```python
# find the mode of the 'Outlet_Size' column
mode(data['Outlet_Size'])
```

```python
mode(data['Outlet_Size']).mode[0]
```

```python
#Filling all the null values in the column 'Outlet_Size' with the mode of the column

data['Outlet_Size'].fillna(mode(data['Outlet_Size']).mode[0], inplace=True)

data.head(10)
```

```python
# noe checking the changes made
data.isnull().sum()
```
